```python
import zipfile
import os

path = '../data'
for f in os.listdir(path):
    if f.endswith('.zip'):
        with zipfile.ZipFile(os.path.join(path, f), 'r') as z:
            z.extractall(path)
            print(f'Extracted: {f}')
```
